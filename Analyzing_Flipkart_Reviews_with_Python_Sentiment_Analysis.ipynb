#**Analyzing Flipkart Reviews with Python: A Sentiment Analysis**

**Flipkart Reviews Sentiment Analysis using Python:**

For this sentiment analysis task, we will work with a dataset downloaded from Kaggle. Let's begin by importing the necessary Python libraries and loading the dataset:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/flipkart_reviews.csv")
print(data.head())
```

**Data Overview:**

The dataset contains three columns. To ensure data quality, let's check for any missing values:

```python
print(data.isnull().sum())
```

Fortunately, the dataset does not have any missing values, so we can proceed. For sentiment analysis, we need to clean and prepare the reviews in the "Review" column:

```python
import nltk
import re
nltk.download('stopwords')

stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string

stopword = set(stopwords.words('english'))

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text = " ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text = " ".join(text)
    return text

data["Review"] = data["Review"].apply(clean)
```

**Sentiment Analysis of Flipkart Reviews:**

To understand how customers rate products on Flipkart, let's visualize the distribution of ratings:

```python
import plotly.express as px

ratings = data["Rating"].value_counts()
numbers = ratings.index
quantity = ratings.values

figure = px.pie(data,
             values=quantity,
             names=numbers,
             hole=0.5)
figure.show()
```

From the pie chart, we observe that 60% of reviewers have given products a 5-star rating on Flipkart.

Word Cloud Analysis:

To gain insights into the most frequently used words in reviews, we'll create a word cloud visualization:

```python
text = " ".join(i for i in data.Review)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="white").generate(text)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

**Analyzing Sentiment Scores:**

Now, let's analyze the sentiments of Flipkart reviews by calculating sentiment scores (Positive, Negative, Neutral) for each review:

```python
nltk.download('vader_lexicon')

sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["Review"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["Review"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["Review"]]
data = data[["Review", "Positive", "Negative", "Neutral"]]
print(data.head())
```

**Reviewing Sentiment Analysis Results:**

Let's review the overall sentiment scores based on Positive, Negative, and Neutral sentiments:

```python
x = sum(data["Positive"])
y = sum(data["Negative"])
z = sum(data["Neutral"])

def sentiment_score(a, b, c):
    if (a > b) and (a > c):
        print("Positive üòä ")
    elif (b > a) and (b > c):
        print("Negative üò† ")
    else:
        print("Neutral üôÇ ")

sentiment_score(x, y, z)

print("Positive: ", x)
print("Negative: ", y)
print("Neutral: ", z)
```
